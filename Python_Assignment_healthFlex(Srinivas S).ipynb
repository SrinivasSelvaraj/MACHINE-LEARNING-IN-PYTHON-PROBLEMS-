```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

file_path = '/content/Python Algo Assignment  - Sheet1.csv'
sensor_data = pd.read_csv(file_path)

acc_x = sensor_data['Acceleration X(g)'] * 9.81
acc_y = sensor_data['Acceleration Y(g)'] * 9.81
acc_z = sensor_data['Acceleration Z(g)'] * 9.81

dt = 1 / 20

vel_x = cumtrapz(acc_x, dx=dt, initial=0)
vel_y = cumtrapz(acc_y, dx=dt, initial=0)
vel_z = cumtrapz(acc_z, dx=dt, initial=0)

disp_x = cumtrapz(vel_x, dx=dt, initial=0)
disp_y = cumtrapz(vel_y, dx=dt, initial=0)
disp_z = cumtrapz(vel_z, dx=dt, initial=0)

dist_x = np.sum(np.abs(np.diff(disp_x)))
dist_y = np.sum(np.abs(np.diff(disp_y)))
dist_z = np.sum(np.abs(np.diff(disp_z)))
total_dist = dist_x + dist_y + dist_z

print(f'Distance X: {dist_x:.2f} m')
print(f'Distance Y: {dist_y:.2f} m')
print(f'Distance Z: {dist_z:.2f} m')
print(f'Total Distance: {total_dist:.2f} m')

plt.figure(figsize=(10, 5))
plt.plot(disp_x, label='Displacement X')
plt.plot(disp_y, label='Displacement Y')
plt.plot(disp_z, label='Displacement Z')
plt.xlabel('Time Steps')
plt.ylabel('Displacement (m)')
plt.legend()
plt.title('Displacement Over Time')
plt.show()
```
