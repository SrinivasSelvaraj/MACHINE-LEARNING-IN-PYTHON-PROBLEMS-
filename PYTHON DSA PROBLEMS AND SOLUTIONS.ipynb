```python
#1
import pandas as pd
df=pd.read_csv('/content/train.csv',index_col='Loan_ID')
df.head()
```

```python
#2
entries=df.shape[0] #row entry
col=df.shape[1] # column entry
print("the total number of entries is {}".format(entries))
print("the total number of column is {}".format(col))
print(df['Gender'].value_counts())
```

```python
#3
avg_income=df['ApplicantIncome'].mean()
print('the average income is {}'.format(avg_income))
female=df[df['Gender']=='Female']
f=female[female['ApplicantIncome']>avg_income][['Gender','ApplicantIncome']]
print('female applicants with income higher than average {}'.format(len(f)))
print(f)
```

```python
#4
df['Loan_Status'].value_counts()
yes=df['Loan_Status'].value_counts()[0]
total =df.shape[0]
print("percentage of loan approved is",yes/total *100)
```

```python
#6
yes=df['Loan_Status'].value_counts()[0]
total=len(df[df['Credit_History']==1])
print("percentage of loans approved is :",(yes/total)*100)
No=df['Loan_Status'].value_counts()[1]
No/total*100
print('Percentage of loan rejected :',(No/total)*100)
```
