```python
import pandas as pd
data = pd.read_csv("/content/MBA Salary.csv")
data.info()
data.shape
```

```python
import numpy as np
X = np.array(data["Percentage in Grade 10"])
Y = np.array(data["Salary"])
```

```python
from sklearn.model_selection import train_test_split

X_train, x_test, Y_train, y_test = train_test_split(X, Y , train_size = 0.8, random_state = 42)
```

```python
def gradient_descent(x,y):
  current_m = current_b = 0
  iterations = 100
  learning_rate = 0.0001
  n = len(x)

  for i in range(iterations):
    predicted_y = current_m * x + current_b
    cost = (1/n) * sum([val ** 2 for val in (y - predicted_y)])
    md = -(2/n) * sum(x * (y - predicted_y))
    bd = -(2/n) * sum(y - predicted_y)

    current_m -= learning_rate * md
    current_b -= learning_rate * bd
    print(f"m is {current_m}, b is {current_b} cost is {cost} iterations is {i}")

  return (current_m, current_b)

x= np.array(X_train)
y = np.array(Y_train)

optimal_parameters = gradient_descent(X_train, Y_train)
print(f"The best values of slope and intercept are {optimal_parameters}")
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

x_test = np.array(x_test)

predict_y = optimal_parameters[0]*x_test + optimal_parameters[1]

print(predict_y)
```

```python
print(f"R2 Score = {np.abs(r2_score(y_test,predict_y))}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, predict_y))}")
```
